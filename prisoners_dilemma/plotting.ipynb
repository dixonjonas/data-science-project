```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the CSV file into a pandas DataFrame
df = pd.read_csv('prisoners_dilemma_results.csv')

# Count the cooperations for each prisoner
prisoner_1_cooperate_count = (df['Prisoner 1 Response'] == 'Cooperate').sum()
prisoner_2_cooperate_count = (df['Prisoner 2 Response'] == 'Cooperate').sum()

# Plotting the cooperations
plt.bar(['Standard LLM personality', 'Ranomized big-five personalities'], [prisoner_1_cooperate_count, prisoner_2_cooperate_count], color=['blue', 'green'])
plt.xlabel('Prisoners')
plt.ylabel('Number of Cooperations')
plt.title('Cooperation Count (out of 100)')
plt.show()
```
