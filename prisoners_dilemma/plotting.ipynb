```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Load the results from the CSV file
data = pd.read_csv('prisoners_dilemma_results.csv')

#Extracting 'Cooperate' and 'Betray'
for i, decisions in enumerate(data['Prisoner 1 Response']):
    if 'cooperate' in decisions:
        data.iloc[i,1] = 'Cooperate'
    elif 'betray' in decisions:
        data.iloc[i,1] = 'Betray'
for i, decisions in enumerate(data['Prisoner 2 Response']):
    if 'cooperate' in decisions:
        data.iloc[i,2] = 'Cooperate'
    elif 'betray' in decisions:
        data.iloc[i,2] = 'Betray'

#Extract the rounds and decisions
rounds = data['Game']
prisoner1_decisions = data['Prisoner 1 Response']
prisoner2_decisions = data['Prisoner 2 Response']

#Map decisions to numerical values (Cooperate = 1, Betray = 0)
decision_mapping = {'Cooperate': 1, 'Betray': 0}
prisoner1_mapped = prisoner1_decisions.map(decision_mapping)
prisoner2_mapped = prisoner2_decisions.map(decision_mapping)

#Plot Prisoner 1's decisions
plt.plot(rounds, prisoner1_mapped, label='Prisoner 1', marker='o', linestyle='-')

#Plot Prisoner 2's decisions
plt.plot(rounds, prisoner2_mapped, label='Prisoner 2', marker='o', linestyle='--')

#Add labels and title
plt.xlabel('Rounds')
plt.ylabel('Decision (1 = Cooperate, 0 = Betray)')
plt.title('Prisoner Decisions Over Time')
plt.legend()

#Show the plot
plt.show()
```
